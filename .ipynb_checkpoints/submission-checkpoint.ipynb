```python
import pandas as pd
import numpy as np
```

```python
from scipy.special import expit
```

```python
from sklearn.metrics import log_loss
```

```python
import rating
import lightgbm as lgb
```

```python
import shap
shap.initjs()
```

### Clean data

```python
def pull_data(train_path, test_path):
    train= pd.read_csv(train_path,parse_dates=['date'])
    test= pd.read_csv(test_path,parse_dates=['date'])
    return train, test
```

```python
train, test = pull_data("https://competition.quantbet.com/data/training.csv", 
                        "https://competition.quantbet.com/data/test.csv")
```

```python
def clean_training_data(train):
    train['gd_ft'] = train['full_time_home_goals'] - train['full_time_away_goals']
    train['gd_ht'] = train['half_time_home_goals'] - train['half_time_away_goals']
    train['outcome_ft'] = train['gd_ft'].apply(win_draw_loss)
    train['outcome_ht'] = train['gd_ht'].apply(win_draw_loss)
    train['is_home'] = 1
    train['gd_ft_exp'] = expit(train['gd_ft'])
    train['gd_ht_exp'] = expit(train['gd_ht'])
    train['division_1'] = train['division'].map({'I1':1,'I2':0})
    train['target'] = train.gd_ft.apply(target_fn)
    train['year'] = train.date.dt.year
    train['month'] = train.date.dt.month
    return train
```

```python
def win_draw_loss(x):
    if x > 0:
        return 1
    elif x == 0:
        return 0.5
    else:
        return -1
```

```python
def target_fn(x):
    if x > 0:
        return 0
    elif x == 0:
        return 1
    else:
        return 2
```

```python
train = clean_training_data(train)
```

### Add elo ratings - various stats

```python
def add_ratings(train, rating_targets):
    res = {}

    train_with_ratings = train.copy()

    for i, target in enumerate(rating_targets):
        print(f'target {target} - {(i+1) / len(rating_targets) *100} %')
        target_rating = rating.ELO(
                fixtures=train.copy(deep=True),
                target=target,
                player_1="home_team",
                player_2="away_team",
            rater_name=target,
            hyperparams={'k_factor':12}
        )
        target_fixtures, target_ratings = target_rating.process_all_fixtures()
        train_with_ratings[f'{target}_e'] =target_fixtures[f'{target}_e']
        res[target] = target_ratings
    return res, train_with_ratings
    
```

```python
RATING_TARGETS = ['outcome_ft','outcome_ht','gd_ft_exp','gd_ht_exp']
```

```python
res, train_with_ratings = add_ratings(train, RATING_TARGETS )
```

```python
def prepare_d_train(train, x_cols):
    X = train.loc[:,x_cols]
    ftgd = train_with_ratings['gd_ft'].append(-1 * train_with_ratings['gd_ft'])
    X_flipped = 1.0-X
    X_flipped[['division_1','year','month']] = -1 * X_flipped[['division_1','year','month']]
    X = X.append(X_flipped)
    y = ftgd.apply(target_fn)
    return  lgb.Dataset(X, y), X, y
```

```python
def gen_elo(rating_1, rating_2):
        q1 = 10.0 ** (rating_1 / 400)
        q2 = 10.0 ** (rating_2 / 400)
        return q1 / (q1 + q2)
```

```python
def prepare_X_test(test,ratings,x_cols):
    test['is_home'] = 1
    test['division_1'] = test['division'].map({'I1':1,'I2':0})
    test['year'] = test.date.dt.year
    test['month'] = test.date.dt.month
    for r in ratings.keys():
        test[f'{r}_e'] = np.nan
        for i, row in test.iterrows():
            home_rating = ratings[r][row.home_team]
            away_rating = ratings[r][row.away_team]
            test.loc[i,f'{r}_e'] = gen_elo(home_rating,away_rating)
    return test[x_cols]
```

```python
X_COLS= ['is_home','outcome_ft_e', 'outcome_ht_e', 'gd_ft_exp_e',
       'gd_ht_exp_e','division_1','year','month']
```

### Train lightbm modle using all of these stats

```python
d_train, X_train, y_train = prepare_d_train(train_with_ratings, 
                                           X_COLS)
```

```python
X_test = prepare_X_test(test, res, X_COLS)
```

```python
PARAMS = {
          "objective" : "multiclass",
          "num_class" : 3,
          "num_leaves" : 60,
          "max_depth": -1,
          "learning_rate" : 0.1,
          "bagging_fraction" : 0.9,  # subsample
          "feature_fraction" : 0.9,  # colsample_bytree
          "bagging_freq" : 5,        # subsample_freq
          "bagging_seed" : 2018,
          "verbosity" : -1 ,
            'random_state': 1234}
```

```python
cv_result_lgb = lgb.cv(PARAMS,
                           d_train,
                           num_boost_round=100,
                            nfold=3,
                               stratified=True,
                               early_stopping_rounds=50, 
                           verbose_eval=10,
                           show_stdv=True,
                          seed=123,)

num_boost_rounds_lgb = len(cv_result_lgb['multi_logloss-mean'])
print(f'num_boost_rounds_lgb={num_boost_rounds_lgb}')
# train model
model_cv = lgb.LGBMClassifier(n_estimators=num_boost_rounds_lgb, **PARAMS)
model_cv.fit(X_train,y_train)
```

```python
train_pred = model_cv.predict_proba(X_train)
```

```python
test_pred = pd.DataFrame(model_cv.predict_proba(X_test))
test_pred.columns = ['home_win','draw','away_win']



test_pred[['away_win','draw','home_win']].to_csv('submission.csv',index=False)
```

```python
train_ll = log_loss(y_train, train_pred)
baseline = log_loss(y_train,  np.ones((y_train.shape[0],3)) /3)
```

```python
print(f'achieved ll of {train_ll:.3f} on training set vs. {baseline:.3f} baseline')
```

```python
explainer = shap.TreeExplainer(model_cv)
shap_values = explainer.shap_values(X_train)
```

```python
shap.summary_plot(shap_values[0], X_train)
shap.summary_plot(shap_values[1], X_train)
shap.summary_plot(shap_values[2], X_train)
```
